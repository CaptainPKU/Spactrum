```python
# version 3.0 measurement results
import numpy as np
import matplotlib.pyplot as plt

FWHM_Hb = []
R_FeII = []
EW_OIII = []
Log_EW_OIII = []
EW_Hb = []

with open('resultsdataV3.txt', 'r') as file:
    # 跳过第一行
    next(file) 
    # 逐行读取文件
    for line in file:
        # 将每行数据按空格或逗号等分隔符分割成列表
        data = line.strip().split()
        # 检查数据列数是否足够
        if len(data) >= 5:
            # 将每列数据分别添加到对应的列表中
            FWHM_Hb.append(float(data[0]))
            R_FeII.append(float(data[1]))
            EW_OIII.append(float(data[2]))
            Log_EW_OIII.append(float(data[3]))
            EW_Hb.append(float(data[4]))

print("Column 1:", FWHM_Hb)
print("Column 2:", R_FeII)
print("Column 3:", EW_OIII)
print("Column 4:", Log_EW_OIII)
print("Column 5:", EW_Hb)

column1 = np.array(FWHM_Hb)
column2 = np.array(R_FeII)
column3 = np.array(EW_OIII)
column4 = np.array(Log_EW_OIII)
column5 = np.array(EW_Hb)

# 创建二维图像
plt.figure(figsize=(6, 6))
# 绘制散点图
sc = plt.scatter(column2, column1, c=column4, cmap='rainbow')
# 添加颜色条
colorbar = plt.colorbar(sc, orientation='horizontal', pad=0.1, location='top')
colorbar.set_label('log(EW_OIII(angstrom))')
# 设置轴标签
plt.xlabel('R_FeII')
plt.ylabel('FWHM_Hb(km/s)')
plt.ylim(0, 10000)
plt.xlim(0, 3)
# 显示图像
plt.show()

EW_FeII=np.array(R_FeII)*np.array(EW_Hb)
print(EW_FeII)
```

```python
# our sample data involved in Boroson&Green1992
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def read_data(file_path):
    # Define column names
    columns = ["Name", "EW_Hb", "EW_5007", "EW_FeII", "FWHM_Hb", "R_FeII"]
    
    # Initialize an empty list to store the data
    data = []
    
    # Open the file and read line by line
    with open(file_path, 'r', encoding='utf-8') as file:
        # Skip the first line as it contains column names
        next(file)
        
        for line in file:
            # Split the line into parts
            parts = line.split()
            
            # If the line has less than 6 parts, it means some data is missing
            if len(parts) < 6:
                # Fill missing data with -100
                parts.extend([-100] * (6 - len(parts)))
            
            # Convert numerical parts to float, keep the name as string
            parts = [parts[0]] + [float(x) for x in parts[1:]]
            
            # Append the parts to the data list
            data.append(parts)
    
    # Create a DataFrame from the data
    df = pd.DataFrame(data, columns=columns)
    
    # Extract each column into a separate list
    name_list = df["Name"].tolist()
    ew_hb_list = df["EW_Hb"].tolist()
    ew_5007_list = df["EW_5007"].tolist()
    ew_feii_list = df["EW_FeII"].tolist()
    fwhm_hb_list = df["FWHM_Hb"].tolist()
    r_feii_list = df["R_FeII"].tolist()
    
    return name_list, ew_hb_list, ew_5007_list, ew_feii_list, fwhm_hb_list, r_feii_list
def plot_data(r_feii_list, fwhm_hb_list, ew_5007_list):
    # Filter out -100 values
    filtered_r_feii = []
    filtered_fwhm_hb = []
    filtered_ew_5007 = []
    
    for r, fwhm, ew in zip(r_feii_list, fwhm_hb_list, ew_5007_list):
        if r != -100 and fwhm != -100 and ew != -100:
            filtered_r_feii.append(r)
            filtered_fwhm_hb.append(fwhm)
            filtered_ew_5007.append(ew)
    
    # Convert EW_5007 to log scale
    log_ew_5007 = np.log10(filtered_ew_5007)
    
    # Create the scatter plot
  
    sc=plt.scatter(filtered_r_feii, filtered_fwhm_hb, c=log_ew_5007, cmap='rainbow')
    #plt.colorbar(label='log10(EW_5007)')
    plt.colorbar(sc, orientation='horizontal', pad=0.1, location='top')
    plt.xlabel('R_FeII')
    plt.ylabel('FWHM_Hb')
    plt.title('Scatter plot of FWHM_Hb vs R_FeII with color as log10(EW_5007)')
    plt.ylim(0, 10000)
    plt.xlim(0, 3)
    plt.show()

file_path = 'c:/Users/liao/PycharmProjects/pythonProject1/AAA.Graduate work/MeasureV3/resultsV3/data_Boroson1992.txt'
name_list, ew_hb_list, ew_5007_list, ew_feii_list, fwhm_hb_list, r_feii_list = read_data(file_path)
print(name_list)
print(ew_hb_list)
print(ew_5007_list)
print(ew_feii_list)
print(fwhm_hb_list)
print(r_feii_list)


plot_data(r_feii_list, fwhm_hb_list, ew_5007_list)
```

```python
import matplotlib.pyplot as plt

def plot_comparison(x_list, y_list, x_label, y_label):
    # Filter out -100 values
    filtered_x = []
    filtered_y = []
    
    for x, y in zip(x_list, y_list):
        if x != -100 and y != -100:
            filtered_x.append(x)
            filtered_y.append(y)
    
    # Create the scatter plot
    plt.scatter(filtered_x, filtered_y, c='blue')
    plt.plot([min(filtered_x), max(filtered_x)], [min(filtered_x), max(filtered_x)], 'r--')  # y=x line
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(f'Scatter plot of {y_label} vs {x_label}')
    plt.show()



# Plot comparisons 横轴全是Boroson1992的数据，纵轴是我们的数据
plot_comparison(ew_5007_list, EW_OIII, 'ew_5007_list', 'EW_OIII')
plot_comparison(fwhm_hb_list, FWHM_Hb, 'fwhm_hb_list', 'FWHM_Hb')
plot_comparison(r_feii_list, R_FeII, 'r_feii_list', 'R_FeII')
plot_comparison(ew_hb_list, EW_Hb, 'ew_hb_list', 'EW_Hb')
plot_comparison(ew_feii_list, EW_FeII, 'ew_feii_list', 'EW_FeII')
```
